````python
#write a program to check given number is even
num=int(input("please enter a number"))
if (num%2 == 0):
    print(num,"is even number")
````

````python
#write a program to find the given number is odd number or even number
num=int(input("please enter a number"))
if (num%2 == 0):
    print(num,"is even number")
else:
    print(num,"is odd number")
````

````python
# write a program to find the given number is positive or negative or zero?
num=int(input("please enter a number"))
if (num<0):
    print(num,"is negative number")
elif (num>0):
    print(num,"is positive number")
else:
    print(num,"is zero")
````

````python
"""
write a program to find the grands based on the student marks
1. if marks in beween 80 to 100 grade is A
2. If marks in between 60 o 79 grade is B
3. If marks in between 40 o 59 grade is C
4. If marks in between 25 o 39 grade is D
5. If marks in between 0 o 24 grade is Fail
6. If marks other than above ouput should be please check valid marks
"""
marks=int(input("please enter student marks here"))
if (marks>=80 and marks<=100):
          print("grade A")
elif (marks>=60 and marks<=79):
          print("grade B")
elif (marks>=40 and marks<=59):
          print("grade C")
elif (marks>=25 and marks<=39):
          print("grade D")
elif (marks>=0 and marks<=24):
          print("Fail")
else:
    print("please check marks")
````
